```python
# ====================
# TRAINING IN COLAB
# ====================
!pip install tensorflow opencv-python

import tensorflow as tf
import cv2
import numpy as np
from google.colab import files

# Simple autoencoder
model = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(128,128,3)),
    tf.keras.layers.Conv2D(32, 3, activation='relu', padding='same'),
    tf.keras.layers.MaxPooling2D(2),
    tf.keras.layers.Conv2D(16, 3, activation='relu', padding='same'),
    tf.keras.layers.MaxPooling2D(2),
    tf.keras.layers.Conv2D(16, 3, activation='relu', padding='same'),
    tf.keras.layers.UpSampling2D(2),
    tf.keras.layers.Conv2D(32, 3, activation='relu', padding='same'),
    tf.keras.layers.UpSampling2D(2),
    tf.keras.layers.Conv2D(3, 3, activation='sigmoid', padding='same')
])

model.compile(optimizer='adam', loss='mse')

# Create dummy data for demonstration
X_train = np.random.rand(100, 128, 128, 3).astype('float32')

# Train
model.fit(X_train, X_train, epochs=5)

# Calculate threshold
reconstructions = model.predict(X_train)
errors = np.mean((X_train - reconstructions)**2, axis=(1,2,3))
threshold = np.percentile(errors, 95)

# Save
model.save('autoencoder_video_anomaly.keras')

import json
with open('threshold.json', 'w') as f:
    json.dump({'threshold': float(threshold)}, f)

# Download to your PC
files.download('autoencoder_video_anomaly.keras')
files.download('threshold.json')
```
